## **ANN Regression with manually uploaded dataset**

```python
# Import necessary libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Load and preprocess the data
df = pd.read_csv(r"C:\Users\sdsow\Downloads\train.csv")
df.head()
```

```python
# Convert 'datetime' column to datetime type
df["datetime"] = df["datetime"].astype("datetime64")
```

```python
# Extract hour, day , month, year from datetime
df["hour"] = df["datetime"].dt.hour
df["day"] = df["datetime"].dt.day
df["month"] = df["datetime"].dt.month
df["year"] = df["datetime"].dt.year
```

```python
df.head()
```

```python
# Drop 'datetime' column
df = df.drop(["datetime"], axis=1)
```

```python
# Define target variable and features
y = df["count"]
X = df.drop(["count"],axis=1)
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=64)
```

```python
data = (X_train - X_train.mean())/X_train.std()
y.head()
```

```python
X_train
```

```python
# Define and train the model

model = Sequential()
# model.add(Dense(128, activation="relu"))
# model.add(Dense(20, activation="linear"))
# model.add(BatchNormalization())
model.add(Dense(16, activation="linear"))
model.add(Dense(1))
model.compile(loss="mse", optimizer="adam")
```

```python
model.fit(X_train,y_train, validation_split=0.2, epochs = 60, verbose = 1)
```

```python
model.evaluate(X,y)
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print evaluation metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2 score):", r2)
```
